```python
# A basic code for Simpson's Rule 
# integrates a given function using using Simpson's 1/3rd rule
import numpy as np

def simpson_int(f, b, a, N = 50):
    # Define Simpson's formula for a function f (defined using lambda or any other suitable method)
    # with upper limit b and lower limit a and divsion range of 50
    
    
    # Sanity check for N, it should be an even number 
    if N%2 == 1:
        raise ValueError("N must be an even integer.")
    
    delta_x = (b-a)/N
    
    x = np.linspace(a,b,N+1)
    
    y = f(x)
    
    S = delta_x/3 * np.sum(y[0:-1:2] + 4*y[1::2] + y[2::2])
    
    #S = delta_x/3*np.sum(y[0] + y[-1] + 4*y[1::2] + 1*y[2::2])
    return S
```

```python
# Example 1: f(x) = x*2, b = 3, a = 0

simpson_int(lambda x : x**2, 3, 0, 400)
```

```python
# Example 2: f(x) = sin(x), b = pi/2 and a = 0

simpson_int(np.sin, np.pi/2, 0, 200)
```
